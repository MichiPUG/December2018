```r
library(ggplot2)
library(dplyr)
```

```r
#### read datafile
exmpl <- read.csv(file="exampleData.csv")
```

```r
#### examine top lines of the dataset
head(exmpl)
```

```r
options( warn = -1 )

#### plot data along with a smoothed linear regression estimate
ggplot(exmpl, aes(x=weight, y = wander_dist)) + 
  geom_point() +
  geom_smooth(method = 'lm')
```

```r
#### run a linear model regression
#### results are contained in the weight_fit object

weight_fit <- lm(formula = wander_dist ~ weight, data=exmpl)
```

```r
#### display a summary of the regression, including coefficient estimates
#### residuals and fit statistics

summary(weight_fit)
```

```r
##### display the structure of the R-object weight_fit

str(weight_fit)
```

```r
#### display the names of the attributes

names(weight_fit)
```

```r
#### display the coefficient attributes using R syntax

weight_fit$coefficients
```

```r
#### examine the top lines of the residuals

head(weight_fit$residuals)
```

```r
##### examine the top lines of the fitted values, using R syntax
head(weight_fit$fitted.values)
```
